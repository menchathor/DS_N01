```python
# Cargo la base para el análisis y normalizo
# df = pd.read_csv('_pokemon.csv')
### df = df.set_index('#')                                              # Defino indice por la columna "#"
# df['Name'] = df['Name'].str.replace(".*(?=Mega|Hoopa )", "")        # Limpio la columna "Name" por error en nombre
# df['Type 2'].value_counts(dropna=False)                             # Visualizo que la columna "Type 2" contiene valores NaN
# df['Type 2'].fillna(df['Type 1'], inplace=True)                     # Relleno los valores NaN con el valor de la columna "Type 1"
df = pd.read_csv('_clases/_db_pokemon.csv')
df['type2'].value_counts(dropna=False)                              # Visualizo que la columna "type2" contiene valores NaN
df['type2'].fillna(df['type1'], inplace=True)                       # Relleno los valores NaN con el valor de la columna "type1"
# df
```

```python
# Visualizo como se distribuyen los Pokémon por tipo primario y secundario
df1 = df['type1'].value_counts().reset_index().rename(columns={"index":"Type", "type1":"Primary"})
df2 = df['type2'].value_counts().reset_index().rename(columns={"index":"Type", "type2":"Secondary"})
df_type = df1.merge(df2, on='Type', how='left').set_index('Type').sort_index()
# df_type
ax1 = df_type.plot.bar(figsize=(15,5), rot=25, ylabel='Quatity', title='Pokémon by type')


# Visualizo las generaciones por cantidad de pokémon
df_gen = df['generation'].value_counts().reset_index().rename(columns={"index":"Generation", "generation":"Pokémon"}).set_index('Generation').sort_index()
# df_gen
ax2 = df_gen.plot.bar(figsize=(12,6), rot=0)
# mpl.style.use("bmh")
# fig, ax = plt.subplots()
# ax.bar(df_pok['Generación'], df_pok['Cantidad'])
# ax.set_title("Cantidad de Pokemón por Generación")
# ax.set_xlabel("Generación")
# ax.set_ylabel("Cantidad")
```
